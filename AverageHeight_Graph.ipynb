```python
import pandas as pd
import matplotlib.pyplot as plt

def convert_height_to_inches(height_str):
    try:
        feet, inches = height_str.split('-')
        return int(feet) * 12 + int(inches)
    except ValueError:  # Handles cases with malformed data such as "-"
        return None

# Load data from CSV files
men_swimming_diving = pd.read_csv('Men_Swimming_Diving.csv')
women_swimming_diving = pd.read_csv('women_SwimmingAndDiving.csv')
men_volleyball = pd.read_csv('men_volleyball.csv')
women_volleyball = pd.read_csv('women_volleyball.csv')

# Apply the conversion function to each DataFrame
men_swimming_diving['Height'] = men_swimming_diving['Height'].apply(convert_height_to_inches)
women_swimming_diving['Height'] = women_swimming_diving['Height'].apply(convert_height_to_inches)
men_volleyball['Height'] = men_volleyball['Height'].apply(convert_height_to_inches)
women_volleyball['Height'] = women_volleyball['Height'].apply(convert_height_to_inches)

# Calculate the average heights for each category
avg_men_swimming_diving = men_swimming_diving['Height'].mean()
avg_women_swimming_diving = women_swimming_diving['Height'].mean()
avg_men_volleyball = men_volleyball['Height'].mean()
avg_women_volleyball = women_volleyball['Height'].mean()

print("Average Heights by Category")
print(f"Men's Swimming and Diving: {avg_men_swimming_diving:.2f} inches")
print(f"Women's Swimming and Diving: {avg_women_swimming_diving:.2f} inches")
print(f"Men's Volleyball: {avg_men_volleyball:.2f} inches")
print(f"Women's Volleyball: {avg_women_volleyball:.2f} inches")

# Define categories
categories = ['Men\'s Swimming and Diving', 'Women\'s Swimming and Diving', 'Men\'s Volleyball', 'Women\'s Volleyball']

# Plotting the averages
plt.figure(figsize=(12, 8))
plt.bar(categories, [avg_men_swimming_diving, avg_women_swimming_diving, avg_men_volleyball, avg_women_volleyball], 
        color=['blue', 'green', 'red', 'yellow'])
plt.title('Average Height Distribution Across Team Categories')
plt.xlabel('Team Category')
plt.ylabel('Average Height (inches)')
plt.xticks(rotation=70)
plt.tight_layout()
plt.show()
```
